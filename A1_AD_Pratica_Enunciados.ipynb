```python
# Bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

pkgdir = 'dados'
```

**1. Faça a leitura dos dados a partir do arquivo covid_caso.csv disponível no material do curso.**

```python
# COVID - Leitura dos dados
covid = pd.read_csv(f'{pkgdir}/covid_caso.csv', index_col='date', parse_dates=True)

covid.head()
```

**Considere apenas os dados de casos e óbitos diários da cidade de São Paulo. Tome as diferenças para obter dados diários.**


```python
# Vamos trabalhar com dados da cidade de São Paulo 

covidSaoPaulo = covid.loc[(covid['city']=='São Paulo')&(covid['place_type']=='city'),['confirmed','deaths']]

# Ordenar dados pela data em ordem crescente

covidSaoPaulo = covidSaoPaulo.sort_values(by=['date'])

covidSaoPaulo.head()
```

```python
# Tomando as diferenças

y1 = list(np.diff(covidSaoPaulo['confirmed']))
y2 = list(np.diff(covidSaoPaulo['deaths']))
x = covidSaoPaulo.index[1:] # Note o subconjunto de dados em x iniciando em 1 pois interessa as diferenças

covidSaoPaulo = pd.DataFrame({'date':x,'confirmed':y1,'deaths':y2})

covidSaoPaulo = covidSaoPaulo.set_index('date') 

covidSaoPaulo.head()
```

**Salve os dados de São Paulo num arquivo covidSaoPaulo.csv.**

```python
pkgdir = 'dados'
covidSaoPaulo.to_csv(f'{pkgdir}/covidSaoPaulo.csv')
```

```python
# Fazendo a leitura do arquivo csv
pkgdir = 'dados'
covidSaoPaulo = pd.read_csv(f'{pkgdir}/covidSaoPaulo.csv', index_col=0, parse_dates=True)
covidSaoPaulo.head()
```

**Produza um gráfico da séries de casos e mortes diários da cidade de São Paulo desde o primeiro caso.**

```python
covidSaoPaulo.plot();
```

```python
title = 'COVID-19 na cidade de São Paulo'
ylabel = ''
xlabel = 'Data'

ax=covidSaoPaulo[covidSaoPaulo['confirmed']>0].plot(title=title);
ax.autoscale(axis='both');
ax.set(xlabel=xlabel,ylabel=ylabel);

ax.xaxis.grid(True) # Com grades
ax.yaxis.grid(True)
```

**4. Considere o gráfico do log de casos a partir somente do centésimo caso.**

```python
# Gráfico do log dos casos

title = 'Log dos casos na cidade de São Paulo a partir do 100o. caso'
ylabel = 'Log dos Casos'
xlabel = 'Tempo (dias)'

ax=np.log(covidSaoPaulo.loc[covidSaoPaulo['confirmed']>100]['confirmed']).plot(c='red', figsize=(12,6), title=title);
# ax=covidSaoPaulo.loc[covidSaoPaulo['confirmed']>100]['confirmed'].plot(figsize=(12,6), c='brown', ls='-.', logy=True, title=title)
ax.autoscale(axis='both')
ax.set(xlabel=xlabel,ylabel=ylabel)
ax.xaxis.grid(True)
ax.yaxis.grid(True);
```

## Exercício 2

1. Construa gráficos da média móvel simples e exponencialmente ponderada para casos. Utilize janela de 7 dias para a MMS e span=7 para MMEP.

2. Construa gráficos da média móvel simples e exponencialmente ponderada para mortes. Utilize janela de 7 dias para a MMS e span=7 para MMEP.


**Construa gráficos da média móvel simples e exponencialmente ponderada para a variável casos.**

```python
covidSaoPaulo['MMS-7'] = covidSaoPaulo['confirmed'].rolling(window=7).mean()
covidSaoPaulo['MMEP-7'] = covidSaoPaulo['confirmed'].ewm(span=7,adjust=False).mean()
```

```python
covidSaoPaulo[['confirmed','MMS-7']].plot();
```

```python
covidSaoPaulo[['confirmed','MMS-7','MMEP-7']].plot(figsize=(8,6));
```

**Construa gráficos da média móvel simples e exponencialmente ponderada para a variável mortes (deaths).**

```python
covidSaoPaulo['MMS-7'] = covidSaoPaulo['deaths'].rolling(window=7).mean()
covidSaoPaulo['MMEP-7'] = covidSaoPaulo['deaths'].ewm(span=7,adjust=False).mean()
```

```python
covidSaoPaulo[['deaths','MMS-7']].plot();
```

```python
covidSaoPaulo[['deaths','MMS-7','MMEP-7']].plot(figsize=(8,6));
```

## Exercício 3

Faça a decomposição em tendência e sazonalidade, aditiva e multiplicativa, do número de casos utilizando a função seasonal_decompose do módulo statsmodels.

```python
# Fazendo a leitura do arquivo csv
pkgdir = 'dados'
covidSaoPaulo = pd.read_csv(f'{pkgdir}/covidSaoPaulo.csv', index_col=0, parse_dates=True)
covidSaoPaulo.head()
```

```python
covidSaoPaulo.index
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams

result = seasonal_decompose(covidSaoPaulo[covidSaoPaulo['confirmed']>0]['confirmed'], model='multiplicative', period=7)

rcParams['figure.figsize'] = 10, 6

fig = result.plot()

plt.show()
```

```python
result = seasonal_decompose(covidSaoPaulo['confirmed'], model='additive', period=7)

rcParams['figure.figsize'] = 10,6

fig = result.plot()

plt.show()
```

**Na sua opinião, qual o tipo de sazonalidade mais adequada, aditiva ou multiplicativa? Justifique com base nos resíduos.**

A sazonalidade multiplicativa é mais adequada pois a distribuição de resíduos é mais uniforme ao longo do tempo.

Olhando para os dados, há uma flutuação que varia ao longo do tempo, isso também indica que a sazonalidade deveria ser multiplicativa.
