    Whether to use the stepwise algorithm outlined in Hyndman and Khandakar

    can be significantly faster than fitting all (or a ``random`` subset

    the model.

Ref.Morettin&Toloi, p.130:<br>

acf(auto-correlation function):<br>
. AR(p) decai de acordo com exponenciais e/ou senóides amortecidas, infinita em extensão;<br>
. MA(q) tem uma acf finita, no sentido que ela apresenta um corte após o "lag" q;<br>
. ARMA(p,q) tem acf infinita em extensão, a qual decai de acordo com exponenciais e/ou senóides amortecidas após o "lag" q-p.<br>

pacf(partial auto-correlation function):<br>
. AR(p) tem pacf != 0, para k <= p e pacf = 0, para k > p;<br>
. MA(q) tem pacf que se comporta de maneira similar à fac de um processo AR(p) e é denominada por exponenciais e/ou senóides amortecidas;<br>
. ARMA(p,q) tem pacf que se comporta como a pacf de um processo MA(q) puro.<br>
